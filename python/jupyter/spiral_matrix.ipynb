```python
matrix = [[7],[9],[6]]
d = matrix 
r = len(matrix)
c = len(matrix[0])
lt = 0      # right 
rt = 0      # bottom
lb = r -1  # up
rb = c -1  # left


l = []

print(r*c)

n = 0 
while n<(r*c):
    if len(l)==(r*c):
        break
        
    if n<(r*c):
        piv = lt
        for i in range(rb+1-piv):
            l.append(d[lt][piv])
            n = n + 1
            piv = piv + 1
        lt = lt +1
        print("27",n)
        
    if len(l)==(r*c):
        break

    if n<(r*c):
        piv = lt
        for i in range(lb-piv):
            l.append(d[piv][rb])
            n = n + 1
            piv = piv + 1
        print("38",n)
            
    if len(l)==(r*c):
        break
        
    if n<(r*c):
        piv = rb 
        for i in range(rb-lt+1):
            l.append(d[lb][piv] )
            n = n + 1
            piv = piv -1
        print("49",n)
        rb = rb -1   
        
    if len(l)==(r*c):
        break

    if n<(r*c):
        piv = lb
        for i in range(lb-lt+1):
            l.append(d[piv][lt-1])
            n = n + 1
            piv = piv -1
        print("61",n)
        lb = lb -1   

print(l)
```
